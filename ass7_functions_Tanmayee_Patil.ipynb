### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
operations=[ {"op": "add", "b": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]
def adapter(operations):
    for i in operations:
        if i["op"]=="add":
            print(i["a"]+i["b"])
        elif i["op"]=="mul":
            print(i["a"]*i["b"])
        else:
            print(i["a"]/i["b"])
adapter(operations)
        
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
data=[ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
def process_students(data):
    for i in data:
        total_marks=sum(i["marks"])
        if total_marks>=240:
            print("Name:", i["name"],"|","Total:",total_marks,"|","Grade: A")
        elif total_marks>=200:
            print("Name:", i["name"],"|","Total:",total_marks,"|","Grade: B")
        else:
            print("None")
process_students(data)
```

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_discounts(products):
    for i in products:
        discount_price=int(i["price"]-(i["price"]*(i["discount"]/100)))
        print(i["name"], "final:",discount_price)
products=[ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ]
apply_discounts(products)
```

### Write a function average_salary(emp_list) that returns the average salary.

### Write total_inventory_value(items) → sum(price * stock).

### Write search_product(items, name) → return matching product(s).

```python
def search_product(items,name):
    result=[]
    for i in items:
        if name.lower() in i['name'].lower():
            result.append(i)
    print(result)

items = [
    {"id": 1, "name": "Notebook", "price": 50, "stock": 15},
    {"id": 2, "name": "Pen", "price": 10, "stock": 120},
    {"id": 3, "name": "Marker", "price": 30, "stock": 8},
    {"id": 4, "name": "Eraser", "price": 5, "stock": 25},
    {"id": 5, "name": "Stapler", "price": 80, "stock": 12}
]

search_product(items,'pen')
```
