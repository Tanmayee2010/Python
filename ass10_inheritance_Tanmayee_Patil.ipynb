# Python OOP Assignment — Inheritance Examples

```python
# 1) Single Inheritance: Person -> Student
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no

# TODO:
# 1. Create object of Student.
s=Student("Tanmayee",12)
# 2. Print name
print(s.name)
# 3. Print roll_no
print(s.roll_no)
```

```python
# 2) Single Inheritance: Animal -> Dog
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# TODO:
# 1. Create Dog object.
d=Dog()
# 2. Call sound() from parent class.
d.sound()
# 3. Call bark() from child class.
d.bark()
```

```python
# 3) Constructor Inheritance: Vehicle -> Car
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

# TODO:
# 1. Create Car object.
c=Car("Maruti Suzuki","Grand i10")
# 2. Print brand.
print(c.brand)
# 3. Print model.
print(c.model)
```

```python
# 4) Multilevel Inheritance: A -> B -> C
class A:
    def showA(self):
        print("This is A")

class B(A):
    def showB(self):
        print("This is B")

class C(B):
    def showC(self):
        print("This is C")

# TODO:
# 1. Create C object.
obj=C()
# 2. Call showA(), showB(), showC().
obj.showA()
obj.showB()
obj.showC()
```

```python
# 5) Multilevel Inheritance with methods: LivingThing -> Animal -> Dog
class LivingThing:
    def living(self):
        print("I am alive")

class Animal(LivingThing):
    def eat(self):
        print("Eating…")

class Dog(Animal):
    def bark(self):
        print("Barking…")

# TODO:
# 1. Create Dog object.
d=Dog()
# 2. Call living(), eat(), bark().
d.living()
d.eat()
d.bark()
```

```python
# 6) Multiple Inheritance: Father + Mother -> Child
class Father:
    def father_method(self):
        print("Father method")

class Mother:
    def mother_method(self):
        print("Mother method")

class Child(Father, Mother):
    def child_method(self):
        print("Child method")

# TODO:
# 1. Create Child object.
c=Child()
# 2. Call father_method(), mother_method(), child_method().
c.father_method()
c.mother_method()
c.child_method()
```

```python
# 7) Constructor Inheritance: Product -> Laptop
class Product:
    def __init__(self, name):
        self.name = name

class Laptop(Product):
    def __init__(self, name, ram):
        super().__init__(name)
        self.ram = ram

# TODO:
# 1. Create Laptop object.
l=Laptop("HP","8 GB")
# 2. Print name.
print(l.name)
# 3. Print ram.
print(l.ram)
```

```python
# 8) Multilevel with __init__: Country -> State -> City
class Country:
    def __init__(self, cname):
        self.cname = cname

class State(Country):
    def __init__(self, cname, sname):
        super().__init__(cname)
        self.sname = sname

class City(State):
    def __init__(self, cname, sname, cityname):
        super().__init__(cname, sname)
        self.cityname = cityname

# TODO:
# 1. Create City object.
c=City("India","Maharashtra","Kolhapur")
# 2. Print cname, sname, cityname.
print(c.cname)
print(c.sname)
print(c.cityname)
```

```python
# 9) Multiple Inheritance with methods: Teacher + Player -> Student
class Teacher:
    def teach(self):
        print("Teaching…")

class Player:
    def play(self):
        print("Playing…")

class Student(Teacher, Player):
    def study(self):
        print("Studying…")

# TODO:
# 1. Create Student object.
s=Student()
# 2. Call teach(), play(), study().
s.teach()
s.play()
s.study()
```

```python
# 10) Constructor Inheritance: Shape -> Rectangle
class Shape:
    def __init__(self, color):
        self.color = color

class Rectangle(Shape):
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

# TODO:
# 1. Create Rectangle object.
r=Rectangle("Pink",25,20)
# 2. Print color, length, width.
print(r.color)
print(r.length)
print(r.width)
```

```python
# ---------------- CORRECTED CLASS ----------------
class order:
    def __init__(self, data):
        self.student_id = data["student"]["id"]
        self.student_name = data["student"]["name"]
        self.items = data["items"]

        self.subtotal = self.get_subtotal()
        self.discount_price = self.get_discount_price()
        self.tax = self.get_tax()
        self.final_total = self.discount_price + self.tax

    def get_subtotal(self):
        total = 0
        for i in self.items:
            total += i["price"] * i["qty"]
        return total

    def get_discount_price(self):
        if self.subtotal >= 150:
            return self.subtotal * 0.95   # 5% discount
        return self.subtotal              # No discount

    def get_tax(self):
        return self.discount_price * 0.05  # 5% tax

ord = [order(o) for o in orders]

# 1) Subtotal
print("\n--- SUBTOTAL ---")
for o in ord:
    print(o.student_id, o.student_name, o.subtotal)

# 2) Discounted price
print("\n--- DISCOUNT PRICE ---")
for o in ord:
    print(o.student_id, o.student_name, round(o.discount_price, 2))

# 3) Tax
print("\n--- TAX ---")
for o in ord:
    print(o.student_id, o.student_name,
          "Subtotal:", o.subtotal,
          "Tax:", round(o.tax, 2))

# 4) Total revenue
total_revenue = sum(o.final_total for o in ord)
print("\n*** TOTAL REVENUE ***")
print(round(total_revenue, 2))
```
