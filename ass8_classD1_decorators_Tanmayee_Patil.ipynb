**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
# TODO: Write your solution for Q24 below
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    @classmethod
    def from_string(cls, s):
        day, month, year = map(int, s.split("-"))
        return cls(day, month, year)
d = Date.from_string("23-11-2024")
print(d.day,"-",d.month,"-",d.year)
```

**Q25.** Create a class `AppConfig` with class variable `DEFAULT_TIMEOUT` and class method to change it. Show that new instances use updated default.

```python
# TODO: Write your solution for Q25 below
class AppConfig:
    DEFAULT_TIMEOUT = 30 
    def __init__(self, timeout=None):
        self.timeout =AppConfig.DEFAULT_TIMEOUT
    @classmethod
    def set_default_timeout(cls, new_timeout):
        cls.DEFAULT_TIMEOUT = new_timeout
a = AppConfig()
print(a.timeout) 
AppConfig.set_default_timeout(60)
b = AppConfig()
print(b.timeout) 
```

## G. Decorators (27–30)


**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
# TODO: Write your solution for Q27 below
def start_fun(func):
    def wrapper():
        print("Function started")
        return func()
    return wrapper

@start_fun
def hello():
    print("Hello!") 
hello()
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
# TODO: Write your solution for Q28 below
def end_fun(func):
    def wrapper():
        result=func()
        print("Function end")
        return result
    return wrapper

@end_fun
def hello():
    print("Bye!") 
hello()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
# TODO: Write your solution for Q29 below
def show_name(func):
    def wrapper():
        print(func.__name__)
        return func()
    return wrapper
@show_name
def hello():
    print("Hello!")
hello()

```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
# TODO: Write your solution for Q30 below
def deco_start(func):
    def wrapper():
        print("Start")
        return func()
    return wrapper

def deco_end(func):
    def wrapper():
        result = func()
        print("End")
        return result
    return wrapper

@deco_start
@deco_end
def hello():
    print("Hello")
hello()
```
