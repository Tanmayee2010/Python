## Q1: Create a class called Student

Requirements:
- Class variable: `school_name = "ABC International School"`
- Constructor: takes `name` and `marks`
- Instance method: `display_details()`
- Class method: `change_school()`
- Static method: `is_pass(marks)` returns "Pass" if marks â‰¥ 35 else "Fail"


```python
class Student():
    school_name="ABC International School"
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def display_details(self):
        print("Name:",self.name)
        print("Marks:",self.marks)
        print("School:",Student.school_name)

    @classmethod
    def change_school(cls, new_name):
        cls.school_name=new_name

    @staticmethod
    def is_pass(marks):
        if marks>=35:
            return 'Pass'
        else:
            return 'Fail'

s1 = Student("Tanmayee", 80)
s1.display_details()
print(Student.is_pass(s1.marks))
Student.change_school("XYZ Public School")
s1.display_details()
```

## Q2: Create a class Calculator

Requirements:
- Class variable: `count_operations = 0`
- Instance method: `add(a, b)`
- Class method: `increment_count()`
- Static method: `info()`


```python
class Calculator:
    count_operations=0
    def add(self,a,b):
        Calculator.count_operations+=1
        return a+b
    
    @classmethod
    def increment_count(cls):
        cls.count_operations+=1

    @staticmethod
    def info():
        return "This is example of calculator"
c1=Calculator()
print("Sum:",c1.add(2,3))
print("Operations Count:",Calculator.count_operations)
Calculator.increment_count()
print("Operations Count after Incrementing:",Calculator.count_operations)
print(Calculator.info())
```

## Q3: Write a class BankAccount

Requirements:
- Class variable: `bank_name = "SBI"`
- Constructor: `name`, `balance`
- Instance method: `deposit(amount)`
- Class method: `change_bank(new_name)`
- Static method: `validate_amount(amount)`


```python
class BankAccount:
    bank_name="SBI"
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self, amount):
        if BankAccount.validate_amount(amount):
            self.balance+=amount
            print(f"Deposit successful! New balance: {self.balance}")
        print("Name:", self.name)
        print("Balance:", self.balance)
        print("Bank Name:", BankAccount.bank_name)
    @classmethod
    def change_bank(cls,new_name):
        cls.bank_name=new_name
    
    @staticmethod
    def validate_amount(amount):
        if amount>0:
            return True
        else:
            print("Invalid amount! Must be greater than 0.")
            return False

b1=BankAccount("tanmayee",800000)
b1.deposit(5000)
print(BankAccount.validate_amount(1000))
BankAccount.change_bank("BOB")
b1.deposit(2000)
```
